```python
%matplotlib inline

import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
```

```python
ax = sns.heatmap(uniform_data, vmin=0, vmax=5)
```

```python
ax = sns.heatmap(uniform_data, robust=True)
```

```python
normal_data = np.random.randn(10, 12)
ax = sns.heatmap(normal_data)
```

```python
ax = sns.heatmap(normal_data, robust=True)
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
#two categorical variable and a continuous variable
ax = sns.heatmap(flights, annot=True, fmt="d")
```

```python
data = np.random.randn(50, 20)
ax = sns.heatmap(data, xticklabels=2, yticklabels=False)
```

```python
ax = sns.heatmap(data, xticklabels=2, yticklabels=False, square=True)
```

```python
ax = sns.heatmap(data, xticklabels=2, yticklabels=False, cbar=False)
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
     ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```
