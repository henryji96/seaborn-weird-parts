```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
ax = sns.boxplot(x=tips["total_bill"])
```

```python
#add confidence interval for the median
ax = sns.boxplot(x=tips["total_bill"], notch=True)
```

```python
#whis : float, optional
#Proportion of the IQR past the low and high quartiles to extend the plot whiskers. 
#Points outside this range will be identified as outliers.
ax = sns.boxplot(x=tips["total_bill"], whis=2)
```

```python
#useful for view distribution of data
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```
