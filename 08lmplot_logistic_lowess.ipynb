```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
g = sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", row="sex", col="time", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
```

```python
#confidence interval around them
#Apply this function to each unique value of x and plot the resulting estimate. 
#This is useful when x is a discrete variable.
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
sns.lmplot(x="size", y="tip", data=tips,
           x_estimator=np.mean, x_ci=50, ci=90);
```

```python
#turn continuous datapoints to discrete
#maybe not so much useful
g = sns.lmplot(x="total_bill", y="tip", data=tips, x_bins=5)
```

```python
anscombe = sns.load_dataset("anscombe")
anscombe.tail()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#change order
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#robust regression, neglect the outliers
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});
```

```python
#
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
print tips.head()

sns.lmplot(x="total_bill", y="big_tip", data=tips,
           y_jitter=.03);
```

```python
#logistic regression
#assume that y is a binary variable and use statsmodels to estimate a logistic regression model.
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03);
```

```python
#estimate a nonparametric lowess model (locally weighted linear regression).
sns.lmplot(x="total_bill", y="tip", data=tips,
           lowess=True);
```

```python
#estimate a linear regression of the form y ~ log(x)
#but plot the scatterplot and regression model in the input space
sns.lmplot(x="total_bill", y="tip", data=tips,
           logx=True);
```

```python
#stop the lm plot at the beginning and the end of the dataset
sns.lmplot(x="total_bill", y="tip", data=tips,
           logx=True, truncate=True);
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,
           logx=True, truncate=True, x_partial="size");
```

```python
#x_partial
#regress out confounding variables 
#strings in data or matrices
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03, x_partial="size");
```
