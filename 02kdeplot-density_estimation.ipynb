```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
ax = sns.kdeplot(tips.total_bill)
```

```python
ax = sns.kdeplot(tips.total_bill, shade=True)
```

```python
#2D Gaussian
ax = sns.kdeplot(tips.total_bill, tips.tip)
```

```python
ax = sns.kdeplot(tips.total_bill, tips.tip, shade=True)
```

```python
#2D density estimation: gaussian kernal
ax = sns.kdeplot(tips.total_bill, tips.tip, n_levels=20)
```

```python
#kernal
#bw:bandwidth 
ax = sns.kdeplot(tips.total_bill, kernel='cos', bw='silverman')
```

```python
#scott is wider:good for things that are normal
ax = sns.kdeplot(tips.total_bill, kernel='epa', bw='scott')
```

```python
#cumulative dencity function
ax = sns.kdeplot(tips.total_bill, cumulative=True)
```

```python
#Number of discrete points in the evaluation grid.
ax = sns.kdeplot(tips.total_bill, gridsize=5)
```

```python
ax = sns.kdeplot(tips.total_bill, cut=0)
```

```python
ax = sns.kdeplot(tips.total_bill, clip=(0,30))
#only take data points between 0:30, exclude the outliers
```
