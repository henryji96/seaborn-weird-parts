```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
g = sns.jointplot(x="total_bill", y="tip", data=tips)
#joint distribution
#marginal distribution
#p value:significance of the corr
```

```python
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg")
#confidence bond
#kde on marginal distribution
```

```python
g = sns.jointplot("total_bill", "tip", data=tips, kind="hex")
```

```python
g = sns.jointplot("total_bill", "tip", data=tips, kind="kde")
```

```python
from scipy.stats import spearmanr
#pearson:how linear is the relationship
#spearman: if a gos up and b goes up, the they have some corr
g = sns.jointplot("size", "total_bill", data=tips, stat_func=spearmanr)
```

```python
g = (sns.jointplot("total_bill", "tip", data=tips)
    .plot_joint(sns.kdeplot, n_levels=6))
```

```python
g = (sns.jointplot("total_bill", "tip", data=tips)
    .plot_joint(sns.kdeplot, n_levels=6)
    .plot_marginals(sns.rugplot))
```
